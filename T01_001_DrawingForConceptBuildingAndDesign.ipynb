## 2 Why this is useful

We can often quickly express ideas visually. Graphical depictions are good at showing relationships between objects. In physics we use diagrams a lot to illustrate concepts. Knowing how to make a clear diagram of the physics in a situation is often a key step towards solving problems

We must also use drawing to design apparatus and develop technical prototypes. Initially such drawings start out as sketches that we use to explore alternative ideas about how to build something that we need. Then the sketches become more systematic and formal, leading to technical design drawings. Ultimately with the use of computers, we create CAD (computer aided design) drawings and files; these can even be used to operate 3D printers and computer-numerically controlled (CNC) machining equipment to fabricate items.

System block diagrams and symbolic schematics (as in electronics or plumbing) are another key aspect of technical design and construction. These can become essential aspects of documenting scientific work and coordinating efforts with collaborators.

More generally, drawings are used to convey and sell ideas. Product design relies on carefully crafted drawings that combine technical accuracy with visual appeal. This also applies to architectural drawing. Even in scientific work, the use of drawings to convey and win support for ideas is important when we want to obtain funding and create interest in the public or amongst peers.

## 3 Learning goals

The primary goal is to obtain and demonstrate a working familiarity with a computer application (Inkscape) for creating drawings of the following types.
- quick doodles or freehand sketches for use in note-taking or to record an idea for later use
- drawings that can be used to explain ideas organize solutions of physics problems
- representational drawings to describe a technical system device, or component that we might want to make
    - intial quick freehand sketch
    - more careful drawing from orthogonal views, in cross-section, and/or in perspective
- system block diagrams

If you do not have access to a computer, it will be acceptable to use traditional drawing tools to create the above types of drawing, with a recommendation that you find the opportunity to convert these to a computer format as soon as reasonably possible.

### 8.0 IMPORTANT NOTE
If you do not have a computer capable of running Inkscape, it will be 100% acceptable for you to do the exercises 8.2-8.6 using pencil, paper and other classic drawing tools. It will be good if you can then submit photos of your drawings to Canvas, but if this is also not possible then we will make some other arrangement for you to demonstrate the work you have done. In fact, we will likely do the pencil/paper versions first in lab. Then you can explore using a drawing app - Inkscape - onn your own, developing skills with this app at your own pace.
### 8.1 Work through the online tutorials as much as needed

https://inkscape-manuals.readthedocs.io/en/latest/

https://inkscape.org/learn/tutorials/

Then use Inkscape to do the following. The results do not need to be elaborate: just enough to capture the main aspects of each technique. Save each drawing as an SVG file on your computer AND then either print your drawing to a PDF document or export it as a PNG file (or both, which is what I did). For each drawing, submit the PDF or PNG to Canvas. PDFs are preferred for submission: one option is to collect all of your PNG drawings into an MS Word document, print this collection to a single PDF and then submit this PDF to Canvas.

NOTE: you may do this particular exercise (the doodling) with either a regular pencip or pen or with the computer. If you use a pen or pencil, take a photo to submit to Canvas.
